## Relational data model - defining a primary key
* In the relational model of databases, a primary key is a specific choice of a minimal set of attributes (columns) that uniquely specify a tuple (row) in a relation(table)

A primary key's main features are:

- It must contain a unique value for each row of data.
- It cannot contain null values.

```python
data_path = 'C:/Users/Lenovo123/ads_covid-19/data/raw/COVID-19/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv'
pd_raw = pd.read_csv(data_path)
pd_raw.head()
```

```python
pd_data_base = pd_raw.rename(columns={'Country/Region' : 'country', 
                                     'Province/State': 'state'})
pd_data_base['state']=pd_data_base['state'].fillna('no')  #ensure a sring, important for indexing
```

```python
pd_data_base = pd_data_base.drop(['Lat','Long'], axis = 1) # droped out Lat and Long columns 
pd_data_base.head()
```

```python
test_pd = pd_data_base.set_index(['state', 'country']).T
```

```python
test_pd.columns
```

```python
test_pd.stack(level=[0,1]).reset_index()
```

```python
pd_relational_model = test_pd.stack(level=[0,1]).reset_index().rename(columns = {'level_0': 'date',
                                                                              0:'confirmed'},)
```

```python
pd_relational_model.head()
```

```python
pd_relational_model.dtypes
```

```python
pd_relational_model['date']= pd_relational_model.date.astype('datetime64[ns]')
```

```python
pd_relational_model.dtypes
```

```python
pd_relational_model.to_csv('C:/Users/Lenovo123/ads_covid-19/data/processed/COVID_relational_confirmed.csv', sep= ';')
```
